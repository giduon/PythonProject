```python
myindex02 = ['에스프레소', '아메리카노', '마끼야또', '카푸치노', '카페모카']
mycolumns02 = ['컴포즈커피', '스타벅스', '이디야커피', '메가커피', '투썸플레이스']
mylist02 =list(10 * onedata for onedata in range(1, 26))
print(mylist02)
```

```python
import numpy as np

mydata02 = np.reshape(mylist02, (5, 5)) # 5행 5열로 형상 변경
mydata02
```

```python
myframe02 = pd.DataFrame(mydata02, index=myindex02, columns=mycolumns02)
myframe02
```

```python
result02 = myframe02.iloc[1] # 1번째 행 추출
print(result02)
print(f'\n타입 확인 : {type(result02)}')
```

```python
result02 = myframe02.iloc[[1,3]] # 몇개의 행 추출
print(result02)

print(f'\n타입 확인 : {type(result02)}')
```
